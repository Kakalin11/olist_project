```python
products=pd.read_csv('olist_products_dataset.csv')
geolocation = pd.read_csv('olist_geolocation_dataset.csv')
sellers = pd.read_csv('olist_sellers_dataset.csv')
customers = pd.read_csv('olist_customers_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')
order_payments = pd.read_csv('olist_order_payments_dataset.csv')
order_reviews = pd.read_csv('olist_order_reviews_dataset.csv')

```

```python
products = products.fillna('NULL')
geolocation = geolocation.fillna('NULL')
sellers = sellers.fillna('NULL')
customers = customers.fillna('NULL')
orders = orders.fillna('NULL')
order_items = order_items.fillna('NULL')
order_payments = order_payments.fillna('NULL')
order_reviews = order_reviews.fillna('NULL')
```

```python
table_list= [products,geolocation,sellers,customers,orders,order_items,order_payments,order_reviews]
```

```python
def get_df_name(table):
    for table in table_list:
        name =[x for x in globals() if globals()[x] is table][0]
        return name
```

```python
for k in table_list:
    insert_query = "INSERT INTO %s VALUES "%get_df_name(k)
    for i in range(k.shape[0]):
        insert_query+="("
        for j in range(k.shape[1]):
            if k.values[i][j]== 'NULL':
                insert_query+=str(k.values[i][j])+','
            else:
                insert_query+="'"
                insert_query+=str(k.values[i][j])
                insert_query+="',"
        insert_query=insert_query[0:-1]+'),'
    insert_query = insert_query[:-1]+';'
    cursor.execute(insert_query)
    conn.commit()
```
